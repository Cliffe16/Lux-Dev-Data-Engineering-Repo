 Part 1: Data Cleaning in Python (10 Questions)


Load all CSV files into pandas and preview the first 5 rows of each.


```python
#Import libraries
import pandas as pd
```

```python
#Load the data
brands_df = pd.read_csv('bike_store_data/brands.csv')
categories_df = pd.read_csv('bike_store_data/categories.csv')
customers_df = pd.read_csv('bike_store_data/customers.csv')
order_items_df = pd.read_csv('bike_store_data/order_items.csv')
orders_df = pd.read_csv('bike_store_data/orders.csv')
products_df = pd.read_csv('bike_store_data/products.csv')
staffs_df = pd.read_csv('bike_store_data/staffs.csv')
stocks_df = pd.read_csv('bike_store_data/stocks.csv')
stores_df = pd.read_csv('bike_store_data/stores.csv')
```

```python
#preview first five rows of each dataframe
print("Brands DataFrame:")
print(brands_df.head(5))
print("\nCategories DataFrame:")
print(categories_df.head(5))
print("\nCustomers DataFrame:")
print(customers_df.head(5))
print("\nOrder Items DataFrame:")
print(order_items_df.head(5))
print("\nOrders DataFrame:")
print(orders_df.head(5))
print("\nProducts DataFrame:")
print(products_df.head(5))
print("\nStaffs DataFrame:")
print(staffs_df.head(5))
print("\nStocks DataFrame:")
print(stocks_df.head(5))
print("\nStores DataFrame:")
print(stores_df.head(5))
```

Check and display the number of missing values per column.


```python
print("Brands:")
print(brands_df.isnull().sum())
print("\nCategories:")
print(categories_df.isnull().sum())
print("\nCustomers:")
print(customers_df.isnull().sum())
print("\nOrder Items:")
print(order_items_df.isnull().sum())
print("\nOrders:")
print(orders_df.isnull().sum())
print("\nProducts:")
print(products_df.isnull().sum())
print("\nStaffs:")
print(staffs_df.isnull().sum())
print("\nStocks:")
print(stocks_df.isnull().sum())
print("\nStores:")
print(stores_df.isnull().sum())
```

Rename all columns to lowercase and replace spaces with underscores.
 


```python
brands_df.columns = brands_df.columns.str.lower().str.replace(' ', '_')
categories_df.columns = categories_df.columns.str.lower().str.replace(' ', '_')
customers_df.columns = customers_df.columns.str.lower().str.replace(' ', '_')
order_items_df.columns = order_items_df.columns.str.lower().str.replace(' ', '_')
orders_df.columns = orders_df.columns.str.lower() .str.replace(' ', "_")
products_df.columns = products_df.columns.str.lower().str.replace(' ', '_')
staffs_df.columns = staffs_df.columns.str.lower().str.replace(' ', '_')
stocks_df.columns = stocks_df.columns.str.lower().str.replace(' ', '_')
stores_df.columns = stores_df.columns.str.lower().str.replace(' ', '_')
```

Convert date columns (like order_date, shipped_date) to datetime objects.


```python
orders_df.info()

orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])
orders_df['required_date'] = pd.to_datetime(orders_df['required_date'])
orders_df['shipped_date'] = pd.to_datetime(orders_df['shipped_date'])

orders_df.info()
```

Remove duplicate rows from each dataset.


```python
#Identify dupliacte rows
print("Brands:")
print(brands_df.duplicated().sum())
print("\nCategories:")
print(categories_df.duplicated().sum())
print("\nCustomers:")
print(customers_df.duplicated().sum())
print("\nOrder Items:")
print(order_items_df.duplicated().sum())
print("\nOrders:")
print(orders_df.duplicated().sum())
print("\nProducts:")
print(products_df.duplicated().sum())
print("\nStaffs:")
print(staffs_df.duplicated().sum())
print("\nStocks:")
print(stocks_df.duplicated().sum())
print("\nStores:")
print(stores_df.duplicated().sum())
```

Standardize text columns like city or state names (title case, trimmed).

```python
#Identify datatypes
print("Brands:")
print(brands_df.dtypes)
print("\nCategories:")
print(categories_df.dtypes)
print("\nCustomers:")
print(customers_df.dtypes)
print("\nOrder Items:")
print(order_items_df.dtypes)
print("\nOrders:")
print(orders_df.dtypes)
print("\nProducts:")
print(products_df.dtypes)
print("\nStaffs:")
print(staffs_df.dtypes)
print("\nStocks:")
print(stocks_df.dtypes)
print("\nStores:")
print(stores_df.dtypes)

customers_df['city'] = customers_df['city'].str.strip().str.title()
customers_df['state'] = customers_df['state'].str.strip().str.title()
stores_df['city'] = stores_df['city'].str.strip().str.title()
stores_df['state'] = stores_df['state'].str.strip().str.title()
```

```python
customers_df['phone'] = customers_df['phone'].fillna('5551234')

print("\nCustomers:")
print(customers_df.isnull().sum())
```

Combine first and last name columns into a single full_name column.


```python
customers_df["full_name" ] = customers_df['first_name'] + " " + customers_df['last_name']
```

Export cleaned dataframes back to CSV for upload.

```python
brands_df.to_csv('brands_df.csv')
categories_df.to_csv('categories_df.csv')
customers_df.to_csv('customers_df.csv')
order_items_df.to_csv('order_items_df.csv')
orders_df.to_csv('orders_df.csv')
products_df.to_csv('products_df.csv')
staffs_df.to_csv('staffs_df.csv')
stocks_df.to_csv('stocks_df.csv')
stores_df.to_csv('stores_df.csv')
```
