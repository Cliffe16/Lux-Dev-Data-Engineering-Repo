**Write Python code to create all tables in PostgreSQL using SQLAlchemy.**



```python
from sqlalchemy import create_engine, Column, Integer, String, ForeignKey, Date, Float
from sqlalchemy.orm import declarative_base     

#Setup connection engine
engine = create_engine('postgresql://cliffe:BLOOMberg411**@localhost/bike_store')

#Create Base class for all tables
Base = declarative_base()

#Define schemas
class Brand(Base):
    __tablename__ = 'brands'
    
    brand_id = Column(Integer, primary_key=True)
    brand_name = Column(String(25))
    
    def __repr__(self):
        return f"<Brand(name='{self.brand_name}')>"
    
class Category(Base):
    __tablename__ = 'categories'
    
    category_id = Column(Integer, primary_key=True)
    category_name = Column(String(100))
    
    def __repr__(self):
        return f"<Category(name='{self.category_name}')>"
    
class Customer(Base):
    __tablename__ = 'customers'
    
    customer_id = Column(Integer, primary_key = True)
    full_name = Column(String(50))
    first_name  = Column(String(20))
    last_name  = Column(String(20)) 
    phone = Column(String(15))
    email = Column(String(100))
    street = Column(String(100))
    city = Column(String(25))
    state = Column(String(25))
    zip_code = Column(String(10)) 
    
    def __repr__(self):
        return f"<Customer(name='{self.full_name}')>"
    
class Order_Item(Base):
    __tablename__ = 'order_items'
    
    order_id = Column(Integer, ForeignKey('orders.order_id'), primary_key=True)
    item_id = Column(Integer, primary_key=True)        
    product_id = Column(Integer, ForeignKey('products.product_id'))    
    quantity = Column(Integer)
    list_price = Column(Float)
    discount = Column(Float)
    
class Order(Base):
    __tablename__ = 'orders'
    
    order_id = Column(Integer, primary_key=True)
    customer_id = Column(Integer, ForeignKey('customers.customer_id'))               
    order_status = Column(Integer)
    order_date = Column(Date)      
    required_date = Column(Date)
    shipped_date = Column(Date)     
    store_id = Column(Integer, ForeignKey('stores.store_id'))                  
    staff_id  = Column(Integer, ForeignKey('staffs.staff_id'))                
    
class Product(Base):
    __tablename__ = 'products'
    
    product_id = Column(Integer,primary_key=True)       
    product_name = Column(String(100))    
    brand_id = Column(Integer, ForeignKey('brands.brand_id'))
    category_id = Column(Integer, ForeignKey('categories.category_id'))  
    model_year = Column(Integer)
    list_price = Column(Float)
    
    def __repr__(self):
        return f"<Product(name='{self.product_name}')>"
    
class Staff(Base):
    __tablename__ = 'staffs'
    
    staff_id = Column(Integer, primary_key=True)
    first_name  = Column(String(10))
    last_name  = Column(String(10))    
    email = Column(String(100))
    phone = Column(String(25))
    active = Column(Integer)
    store_id = Column(Integer, ForeignKey('stores.store_id'))       
    manager_id = Column(Float)
    
    def __repr__(self):
       return f"<Staff(name='{self.first_name}')>"
   
class Stock(Base): 
    __tablename__ = 'stocks' 
    
    store_id = Column(Integer, ForeignKey('stores.store_id'), primary_key=True)     
    product_id = Column(Integer, ForeignKey('products.product_id'), primary_key=True )
    quantity = Column(Integer)     

class Store(Base):
    __tablename__ = 'stores'  

    store_id = Column(Integer, primary_key=True)
    store_name = Column(String(255), nullable=False)
    phone = Column(String(25))
    email = Column(String(255))
    street = Column(String(255))
    city = Column(String(255))
    state = Column(String(100))
    zip_code = Column(String(10))

    def __repr__(self):
       return f"<Store(name='{self.store_name}')>"
   
#Call the Base class and it's child classes
Base.metadata.create_all(engine)   
print("Tables created")
```

**Write a Python function to verify all foreign key relationships.**

```python
from sqlalchemy import create_engine, inspect, text

#Define engine
engine = create_engine('postgresql://cliffe:BLOOMberg411**@localhost/bike_store')

def verify_fks(engine):
    #Create an inspector to read db schema
    inspector = inspect(engine)
    
    #Get tables in db
    table_names = inspector.get_table_names()
    
    #Initialize fk orphan counter
    issues_found = 0
    
    with engine.connect() as conn:
        for child_table in table_names:
            #get fks
            fks = inspector.get_foreign_keys(child_table)
            
            for fk in fks:
                child_col = fk['constrained_columns'][0]
                parent_table = fk['referred_table']
                parent_col = fk['referred_columns'][0]
                
                #Count child rows where ID is not in parent
                query = text(f"""
                            SELECT COUNT(*)
                            FROM {child_table} c
                                LEFT JOIN {parent_table} p
                                    ON c.{child_col}=p.{parent_col}
                            WHERE p.{parent_col} IS NULL;
                            """)
                #Run the query
                orphan_count = conn.execute(query).scalar()
                
                if orphan_count > 0:
                    print(f"INTEGRITY ERROR: Found {orphan_count} orphans in '{child_table}' referencing {parent_table}")
                    issues_found += 1
                
                else:
                    pass
    
    if issues_found == 0:
        print("No orphans found")
    else:
        print(f"Found {issues_found} integrity issues")
        
verify_fks(engine)        
           
    
```

**Write a CASE query to label orders as Late, On Time, or Early.**

```python
query = f"""
    SELECT  order_id,
            required_date,
            shipped_date,
            CASE 
                WHEN required_date > shipped_date THEN 'Early'
                WHEN required_date < shipped_date THEN 'Late'
                ELSE 'On Time'
            END AS delivery_status
    FROM orders;
    """

result = pd.read_sql(query, con=engine)    
print(result)
```

**Label customers with Low, Medium, or High order volume.**

```python
query = """
    WITH order_count_per_customer AS (
        SELECT
            customer_id,
            COUNT(order_id) AS order_count
        FROM orders
        GROUP BY customer_id
    )
    SELECT
        c.customer_id,
        c.full_name,
        o.order_count,
        CASE
            WHEN o.order_count <= 2 THEN 'Low'
            WHEN o.order_count BETWEEN 3 AND 5 THEN 'Medium'
            ELSE 'High'
        END AS order_volume
    FROM customers c
        JOIN order_count_per_customer o 
            ON c.customer_id = o.customer_id
    ORDER BY o.order_count DESC;
"""

result = pd.read_sql(query, con=engine)
print(result)
```

**Check whether each product is In Stock or Out of Stock.**

```python
query = """
    SELECT
        p.product_name,
        SUM(s.quantity) AS total_stock,
        CASE
            WHEN SUM(s.quantity) > 0 THEN 'In Stock'
            ELSE 'Out of Stock'
        END AS stock_status
    FROM products p
        JOIN stocks s 
            ON p.product_id = s.product_id
    GROUP BY p.product_name
    ORDER BY total_stock ASC;
"""

result = pd.read_sql(query, con=engine)
print(result)
```

**Add a revenue tier column per store using CASE with SUM.**

```python
query = """
    WITH revenue AS (
        SELECT
            o.store_id,
            SUM(i.quantity * i.list_price * (1.0 - i.discount)) AS total_revenue
        FROM orders o
            JOIN order_items i 
                ON o.order_id = i.order_id
        GROUP BY o.store_id
    )
    SELECT
        s.store_id,
        s.store_name,
        r.total_revenue,
        CASE
            WHEN r.total_revenue <= 1000000.00 THEN 'Low Revenue'
            WHEN r.total_revenue > 1000000.00 AND r.total_revenue <= 5000000.00 THEN 'Medium Revenue'
            ELSE 'High Revenue'
        END AS revenue_tier
    FROM stores s
        JOIN revenue r 
            ON s.store_id = r.store_id
    ORDER BY r.total_revenue DESC;
"""

pd.options.display.float_format = '{:,.2f}'.format

result = pd.read_sql(query, con=engine)
print(result)
```

**Automate running queries and saving results as daily reports.**

```python
#report_generator.py
from datetime import datetime

def generate_daily_report():
    report_dir = 'daily_reports'
    
    #Ensure reports directory exists
    os.makedirs(report_dir, exist_ok=True)
    
    query = f"""
        SELECT
            b.brand_name,
            SUM(i.quantity * i.list_price * (1.0 - i.discount)) AS total_revenue
        FROM orders o
            JOIN order_items i 
                ON o.order_id = i.order_id
                    JOIN products p 
                        ON i.product_id = p.product_id
                            JOIN brands b 
                                ON p.brand_id = b.brand_id
        GROUP BY b.brand_name
        ORDER BY total_revenue DESC;
    """
    
    result = pd.read_sql(query, con=engine)
    
    #Format date and file name
    timestamp = datetime.now().strftime("%Y_%m_%d_%H%M%S")
    filename = os.path.join(report_dir, f"revenue_by_brand_{timestamp}.csv")
    
    #Save report in csv
    result.to_csv(filename, index=False)
    
    print("Daily report successfully saved in csv")
    
if __name__ == '__main__':
    generate_daily_report()    
    
```

```python
#error_logger.py
from sqlalchemy.exc import ProgrammingError, OperationalError 
import logging

#Setup file
log_file = 'sql+python test/error_logs'

#Configure logging to write to a file
logging.basicConfig(
    filename=log_file,
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def log_errors():    
    #faulty query for testing
    query = """
        SELECT
            b.brand_name, 
            SUM(i.quantity * i.list_price * (1.0 - i.discount)) AS total_revenue
        FROM orders o
            JOIN order_items i 
                ON o.order_id = i.order_id
                    JOIN products p 
                        ON i.product_id = p.product_id
                            JOIN brands b 
                                ON p.brand_id = b.brand_id
        ORDER BY total_revenue DESC;
    """
    
    try:
        print("Connecting...")
        result = pd.read_sql(query, con=engine)
        print(f"Query ran successfully. {len(result)} rows found")
        
    except ProgrammingError as e:
        error_msg = f"SQL Programming Error in query: Check syntax/schema. Details: {e.__cause__}"
        
        logging.error(error_msg)
        print(f"Handled SQL Error. Check {log_file} for details.")
        
    except OperationalError as e:
        error_msg = f"Database Connection Error: Check server status or credentials. Details: {e.__cause__}"
        
        logging.critical(error_msg)
        print(f"Handled Connection Error. Check {log_file} for details.")
        
    except Exception as e:
        error_msg = f"Unexpected Python Error: {e}"
        logging.critical(error_msg)
        print(f"Handled Unexpected Error. Check {log_file} for details.")
        
if __name__ == '__main__':
    log_errors()   
    
    
```

**Add a CASE column in that query to label order status.**

```python
query = f"""
    SELECT *,
    CASE
        WHEN o.order_status = 1 THEN 'Pending'
        WHEN o.order_status = 2 THEN 'Processing'
        WHEN o.order_status = 3 THEN 'Rejected'
        ELSE 'Completed'
    END AS order_status        
    FROM orders o
        JOIN customers c
            ON o.customer_id=c.customer_id
                JOIN order_items i
                    ON o.order_id=i.order_id
    """
    
result = pd.read_sql(query, con=engine)    
print(result)    
```
