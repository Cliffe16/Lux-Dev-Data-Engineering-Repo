**Write Python code to create all tables in PostgreSQL using SQLAlchemy.**



```python
from sqlalchemy import create_engine, Column, Integer, String, ForeignKey, Date, Float
from sqlalchemy.orm import declarative_base     

#Setup connection engine
engine = create_engine('postgresql://cliffe:BLOOMberg411**@localhost/bike_store')

#Create Base class for all tables
Base = declarative_base()

#Define schemas
class Brand(Base):
    __tablename__ = 'brands'
    
    brand_id = Column(Integer, primary_key=True)
    brand_name = Column(String(25))
    
    def __repr__(self):
        return f"<Brand(name='{self.brand_name}')>"
    
class Category(Base):
    __tablename__ = 'categories'
    
    category_id = Column(Integer, primary_key=True)
    category_name = Column(String(100))
    
    def __repr__(self):
        return f"<Category(name='{self.category_name}')>"
    
class Customer(Base):
    __tablename__ = 'customers'
    
    customer_id = Column(Integer, primary_key = True)
    full_name = Column(String(50))
    first_name  = Column(String(20))
    last_name  = Column(String(20)) 
    phone = Column(String(15))
    email = Column(String(100))
    street = Column(String(100))
    city = Column(String(25))
    state = Column(String(25))
    zip_code = Column(String(10)) 
    
    def __repr__(self):
        return f"<Customer(name='{self.full_name}')>"
    
class Order_Item(Base):
    __tablename__ = 'order_items'
    
    order_id = Column(Integer, ForeignKey('orders.order_id'), primary_key=True)
    item_id = Column(Integer, primary_key=True)        
    product_id = Column(Integer, ForeignKey('products.product_id'))    
    quantity = Column(Integer)
    list_price = Column(Float)
    discount = Column(Float)
    
class Order(Base):
    __tablename__ = 'orders'
    
    order_id = Column(Integer, primary_key=True)
    customer_id = Column(Integer, ForeignKey('customers.customer_id'))               
    order_status = Column(Integer)
    order_date = Column(Date)      
    required_date = Column(Date)
    shipped_date = Column(Date)     
    store_id = Column(Integer, ForeignKey('stores.store_id'))                  
    staff_id  = Column(Integer, ForeignKey('staffs.staff_id'))                
    
class Product(Base):
    __tablename__ = 'products'
    
    product_id = Column(Integer,primary_key=True)       
    product_name = Column(String(100))    
    brand_id = Column(Integer, ForeignKey('brands.brand_id'))
    category_id = Column(Integer, ForeignKey('categories.category_id'))  
    model_year = Column(Integer)
    list_price = Column(Float)
    
    def __repr__(self):
        return f"<Product(name='{self.product_name}')>"
    
class Staff(Base):
    __tablename__ = 'staffs'
    
    staff_id = Column(Integer, primary_key=True)
    first_name  = Column(String(10))
    last_name  = Column(String(10))    
    email = Column(String(100))
    phone = Column(String(25))
    active = Column(Integer)
    store_id = Column(Integer, ForeignKey('stores.store_id'))       
    manager_id = Column(Float)
    
    def __repr__(self):
       return f"<Staff(name='{self.first_name}')>"
   
class Stock(Base): 
    __tablename__ = 'stocks' 
    
    store_id = Column(Integer, primary_key=True)     
    product_id = Column(Integer, ForeignKey('products.product_id', primary_key=True ))   
    quantity = Column(Integer)     

class Store(Base):
    __tablename__ = 'stores'  

    store_id = Column(Integer, primary_key=True)
    store_name = Column(String(255), nullable=False)
    phone = Column(String(25))
    email = Column(String(255))
    street = Column(String(255))
    city = Column(String(255))
    state = Column(String(100))
    zip_code = Column(String(10))

    def __repr__(self):
       return f"<Store(name='{self.store_name}')>"
   
#Call the Base class and it's child classes
Base.metadata.create_all(engine)   
print("Tables created")
```

**Upload cleaned CSV data into tables using pandas .to_sql().**

```python
#Upload tables with parents first
brands_df.to_sql('brands', con=engine, if_exists='append', index=False)
categories_df.to_sql('categories', con=engine, if_exists='append', index=False)
stores_df.to_sql('stores', con=engine, if_exists='append', index=False)
customers_df.to_sql('customers', con=engine, if_exists='append', index=False)
products_df.to_sql('products', con=engine, if_exists='append', index=False)
staffs_df.to_sql('staffs', con=engine, if_exists='append', index=False)
stocks_df.to_sql('stocks', con=engine, if_exists='append', index=False)
orders_df.to_sql('orders', con=engine, if_exists='append', index=False)
order_items_df.to_sql('order_items', con=engine, if_exists='append', index=False)
```
