```python
class BankAccount:
  def __init__(self, balance):
    self.balance = balance

  def deposit(self, amount):
    self.balance += amount

  def withdraw(self, amount):
    if amount > self.balance:
      print("Insufficient funds.")
    else:
      self.balance -= amount

  def get_balance(self):
    return self.balance

acc = BankAccount(1000)
acc.deposit(500)
print(acc.get_balance())
print(acc.withdraw(2000))
```

```python
class Temperature:
  def cel_fah(self, celsius):
    return (celsius * 9/5) + 32

  def fah_cel(self, fahrenheit):
    return (fahrenheit - 32) * 5/9

converter = Temperature()

f_temp = converter.cel_fah(100)
print(f"100°C is {f_temp}°F")

# 2. Convert 32°F to Celsius
c_temp = converter.fah_cel(32)
print(f"32°F is {c_temp}°C")
```

```python
class ToDoList:
  def __init__(self):
    self.task = {}
    self.flag = ['completed', 'pending']

  def add_task(self, task, flag):
    self.task[task] = flag
    return f"{self.task}"

  def get_task(self, flag):
    search = []

    for task_name, task_flag in self.task.items():
      if task_flag == flag:
        search.append(task_name)
    return search


task = ToDoList()
print(task.add_task("Shopping","Completed"))
print(task.add_task("Laundry","Pending"))

completed_tasks = task.get_task("Completed")
print(f"Completed: {completed_tasks}")
pending_tasks = task.get_task("Completed")
print(f"Pending: {pending_tasks}")



```

```python
class Queue:
  def __init__(self):
    self.queue = []

  def enqueue(self, item):
    self.queue.append(item)

  def dequeue(self):
    return self.queue.pop(0)

  def __str__(self):
    return f"{self.queue}"

my_queue = Queue()
my_queue.enqueue("Car")
my_queue.enqueue("Coffee")
print(my_queue)
my_queue.dequeue()
print(my_queue)
```

```python
import random

class Playlist:
  def __init__(self):
    self.tracks = []

  def add_song(self, title):
    self.tracks.append(title)

  def remove_song(self,title):
    return self.tracks.remove(title)

  def shuffle(self):
    random.shuffle(self.tracks)

  def __str__(self):
    return f"{self.tracks}"

my_playlist = Playlist()
my_playlist.add_song("Song A")
my_playlist.add_song("Song B")
my_playlist.add_song("Song C")
print(f"Original playlist: {my_playlist}")

my_playlist.shuffle()

print(f"Shuffled playlist: {my_playlist}")

my_playlist.remove_song("Song B")

print(f"Edited playlist: {my_playlist}")
```

```python
class Matrix:
  def __init__(self, num1, num2, num3, num4):
    self.matrix = [[num1, num2],[num3, num4]]

  def add_matrix(self, second_matrix):
    sum_tl = self.matrix[0][0] + second_matrix.matrix[0][0]
    sum_tr = self.matrix[0][1] + second_matrix.matrix[0][1]
    sum_bl = self.matrix[1][0] + second_matrix.matrix[1][0]
    sum_br = self.matrix[1][1] + second_matrix.matrix[1][1]

    sum_matrix = [[sum_tl, sum_tr], [sum_bl, sum_br]]

    return sum_matrix

m1 = Matrix(1, 2, 3, 4)
m2 = Matrix(1, 2, 3, 4)
print(m1.add_matrix(m2))
```

```python
class VotingSystem:
  def __init__(self):
    self.ballot = {}

  def add_votes(self, name, no_votes):
    if name in self.ballot:
      self.ballot[name] += no_votes
    else:
      self.ballot[name] = no_votes

  def display(self):
    return self.ballot

ballot_box = VotingSystem()
ballot_box.add_votes("Cliffe", 20)
ballot_box.add_votes("Sam", 20)

print(ballot_box.display())
ballot_box.add_votes("Cliffe", 5)

print(ballot_box.display())
```

```python
import statistics

class DataAnalyzer:
  def __init__(self):
    self.nums = []

  def add(self, num1, num2, num3, num4):
    nums_to_add = num1, num2, num3, num4

    for nums in nums_to_add:
      self.nums.append(nums)

  def min(self):
    return f"Minimum: {min(self.nums)}"

  def max(self):
    return f"Maximum: {max(self.nums)}"

  def mean(self):
    mean_num = sum(self.nums)/len(self.nums)
    return f"Mean: {mean_num}"

  def median(self):
    median_num = statistics.median(self.nums)
    return f"Median: {median_num}"

data = DataAnalyzer()
data.add(52, 95, 37, 10)

print(data.min())
print(data.max())
print(data.mean())
print(data.median())

```
