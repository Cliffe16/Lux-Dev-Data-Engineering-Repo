**Combine extract → transform → load into a single Python script.**

**Write Python code that checks if the dataset file exists before running extract.**

**Write code that logs the number of rows before and after transformation.**

```python
import os

def extract(path):
    raw_data = pd.read_csv(path)
    return raw_data
    
def transform(raw_data):
    print("Number of rows before transformation: ", len(raw_data))
    
    df = raw_data
    clean_data = df.drop_duplicates('title').dropna(subset='title')
    clean_data['title'] = clean_data['title'].str.lower()
    
    print("Number of rows after transformation", len(clean_data))    
    return clean_data
    
def load(clean_data):
    return clean_data.to_csv('clean_data.csv', index=False)

if __name__ == "__main__":
    if os.path.exists('data/netflix_titles.csv'):
        raw_data_flix = extract('data/netflix_titles.csv')
        clean_data_flix = transform(raw_data_flix)
        load(clean_data_flix)
        print("Pipeline ran successfully")
        
    else:
        print("File not found")
```

**Write a function that creates a summary dictionary:**

```python
def summary(df):
    sum_dict = {}
    
    sum_dict['total_movies'] = len(df[df['type'] == 'Movie'])
    sum_dict['total_shows'] = len(df[df['type'] == 'TV Show'])
    
    return sum_dict
```
